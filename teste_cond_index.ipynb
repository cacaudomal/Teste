```python
def calc_girofreq(mi,B):
    """
    Funcao para cálculo da girofrequência ou frequência de ciclotron da
    partícula num ponto.
    
    Parameters
    ----------
    B : FLOAT
        campo magnético em [T]
    mi : FLOAT
        massa do íon ou do elétron [kg]    
    q : FLOAT
        carga do íon/elétron [C]
    
    Returns:
    ----------
    wi : FLOAT
        girofrequencia [Hz] 
    """
    e = -1.602177e-19 #Carga do elétron [C]
    
    wi = np.sqrt(e**2) * np.sqrt(B**2)/mi
    return wi 
```

# Calculando girofrequência angular

```python
def calc_all_girofreq(B):
        me = 9.109389e-31 #Massa do elétron em repouso [kg]
        e = -1.602177e-19 #Carga do elétron [C]
        mi1 = 5.065e-26 #Massa do íon 1 uma mistura de NO+ (75%) e O2+ (25%) (30.5 u.m.a.) [kg]
        mi2 = 2.657e-26 #Massa do íon 2 (O+) [kg] (16 a.m.u)
        
        print("\nCalculating all Girofrequencies...")
        wi1 = calc_girofreq(mi1, B)
        wi2 = calc_girofreq(mi2, B)
        we = calc_girofreq(me, B)
        #print("\n\ncondiono_adachi calc_all_girofreq  wi1: ",wi1,"\nwi2:",wi2,"\nwe5:",we)
       # print('lon: ',long,'lat: ',lat)
        
        agirofreq = pd.concat([we,wi1,wi2], axis = 1, keys = ["we","wi1","wi2"],join='inner')
            
        print("Done")
        return agirofreq
```
